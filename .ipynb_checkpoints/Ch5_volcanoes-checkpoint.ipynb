#### This notebook analyses the surface air temperature from the DAMIP hist-nat run using ACCESS-CM2 from 1850-2020. I generate a time series of the global average surface air temperature, with the goal of using this time series to force my 1D diffusion model.

```python
import cosima_cookbook as cc
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import cmocean as cm
from scipy.optimize import curve_fit
from xhistogram.xarray import histogram

from dask.distributed import Client
import cftime

%matplotlib inline
```

```python
client = Client(n_workers=8)
display(client)
```

```python
tas = xr.open_dataset('/home/561/mp2135/HonoursThesis/Ch5_volcanoes/tas_Amon_ACCESS-CM2_hist-nat_r3i1p1f1_gn_185001-202012.nc').tas
areacella = xr.open_dataset('/home/561/mp2135/HonoursThesis/Ch5_volcanoes/areacella_fx_ACCESS-CM2_hist-nat_r3i1p1f1_gn.nc').areacella
```

```python
tas_weighted_ave = (tas * areacella).sum(dim=('lat', 'lon')) / areacella.sum(dim=('lat', 'lon'))
tas_weighted_ave_ann = tas_weighted_ave.groupby('time.year').mean(dim='time')
tas_weighted_ave_ann_anom = tas_weighted_ave_ann - tas_weighted_ave_ann.mean(dim='year')
```

```python
plt.figure(figsize=(25,7))
tas_weighted_ave_ann_anom.plot()
plt.grid()
plt.xlabel('year', fontsize=16)
plt.ylabel('Temeprature anomaly (K)', fontsize=16)
plt.title('Annually averaged global average SAT anomaly, DAMIP hist-nat ACCESS-CM2', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Linearly interpolate monthly down to 10 day increments to use in the 1D model

tas_weighted_ave_dt10 = tas_weighted_ave.resample(time="1D").interpolate("linear")[::10].dropna('time')
tas_weighted_ave_dt10_anom = tas_weighted_ave_dt10 - tas_weighted_ave_dt10.mean(dim='time')
```
