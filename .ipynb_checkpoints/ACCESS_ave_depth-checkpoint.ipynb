```python
### Generate new session

db = '/scratch/e14/mp2135/access-om2/archive/databases/1deg_jra55_ryf_experiments_abrupt_spinup500.db'
session = cc.database.create_session(db)
```

```python
yr_beg = '2400-01-16'
yr_end = '2400-03-16'

expt = '1deg_jra55_ryf_cont'
var = 'temp'
temp = (cc.querying.getvar(expt, var, session, start_time = yr_beg, end_time = yr_end).mean(dim='time')).load()

temp.values[temp.values > -50] = 1 # Make all non-NaN values = 1
```

```python
yr_beg = '2400-01-16'
yr_end = '2400-03-16'

surface_mask = temp[0,:,:].values # temp has been 'normalised' to be 1 if there's a value and nan otherwise

dzt = cc.querying.getvar(expt, 'dzt', session, start_time = yr_beg, end_time = yr_end).mean(dim='time') * surface_mask
area_t = cc.querying.getvar(expt, 'area_t', session, n = 1) * surface_mask

ocean_vol = dzt * area_t

depth = ocean_vol.sum().values / (area_t).sum().values
```
