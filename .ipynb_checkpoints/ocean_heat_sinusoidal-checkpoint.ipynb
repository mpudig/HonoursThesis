```python
# Fit a linear trend to anomaly

def linear_fit(t, a, b):
    
    return a*t + b

def fit_linear_trend(T):
    
    time_dim = T.dims[0]
    t = T[time_dim].values
    popt, pcov = curve_fit(linear_fit, t, T)
    linear_trend = linear_fit(t, *popt)
    return linear_trend, popt[0]

gat_period2_anom = gat_period2 - gat_cont
gat_period5_anom = gat_period5 - gat_cont
gat_period7_anom = gat_period7 - gat_cont
gat_period10_anom = gat_period10 - gat_cont
gat_period12_anom = gat_period12 - gat_cont
gat_period16_anom = gat_period16 - gat_cont
gat_period20_anom = gat_period20 - gat_cont

# gat_period2_anom_trend = fit_linear_trend(gat_period2_anom)[0]
# gat_period2_anom_trend_gradient = fit_linear_trend(gat_period2_anom)[1]
# gat_period5_anom_trend = fit_linear_trend(gat_period5_anom)[0]
# gat_period5_anom_trend_gradient = fit_linear_trend(gat_period5_anom)[1]
# gat_period10_anom_trend = fit_linear_trend(gat_period10_anom)[0]
# gat_period10_anom_trend_gradient = fit_linear_trend(gat_period10_anom)[1]
# gat_period20_anom_trend = fit_linear_trend(gat_period20_anom)[0]
# gat_period20_anom_trend_gradient = fit_linear_trend(gat_period20_anom)[1]
```

```python
### First, average over window length = period

def average_over_period(T_anom, period):
    
    num_periods = np.int(100 / period)
        
    data = np.array([T_anom[period * i * 365 : period * (i + 1) * 365].mean(dim='time').values for i in range(num_periods)])
    times = np.array([T_anom['time'][int(period * i * 365 + 0.5 * period * 365)].values.tolist() for i in range(num_periods)]) # the .tolist() seems incorrect but it works
    
    return xr.DataArray(data, dims=['time'], coords={'time': times})


gat_period2_anom_period_ave1 = average_over_period(gat_period2_anom, 2)
gat_period5_anom_period_ave1 = average_over_period(gat_period5_anom, 5)
gat_period7_anom_period_ave1 = average_over_period(gat_period7_anom, 7)
gat_period10_anom_period_ave1 = average_over_period(gat_period10_anom, 10)
gat_period12_anom_period_ave1 = average_over_period(gat_period12_anom, 12)
gat_period16_anom_period_ave1 = average_over_period(gat_period16_anom, 16)
gat_period20_anom_period_ave1 = average_over_period(gat_period20_anom, 20)
```

```python
def gaussian_pdf(x, mean, amplitude, standard_deviation):
    return amplitude * np.exp( - (x - mean)**2 / (2*standard_deviation ** 2))

def log_normal_pdf(x, mu, sigma):
    return np.exp(- (np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi))

from scipy.special import gamma
def gamma_pdf(x, alpha, beta):
    return beta**alpha / gamma(alpha) * x**(alpha - 1) * np.exp(- beta * x)
```
